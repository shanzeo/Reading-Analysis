**Background and Motivation:**

Reading is widely recognized as an important factor in personal and intellectual growth, with potential impacts on life outcomes such as education, career success, and even relationships. While many studies suggest that reading enhances cognitive function and contributes to success, the strength and nature of these effects remain unclear.

The motivation for this project is to explore how reading habits relate to various life outcomes, including income, career success, and relationships. **Methodology:**

First, I imported the necessary libraries and loaded the dataset books.csv to examine its first few rows.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv("/Users/shanzeowais/Downloads/books.csv", delimiter=',') 
head = data.head()
head
```

Here, I'm filtering the dataset to focus on 50-year-old participants and visualizing their reading habits. I created a scatter plot to analyze the relationship between the number of books read and income, and a box plot to compare reading habits across different education levels for this age group.

```python
data_50 = data[data['Age'] == 50]

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_50['How many books did you read during last 12months?'], y=data_50['Incomes'], color='purple')
plt.title('Books Read vs Income for 50-Year-Olds')
plt.xlabel('Number of Books Read')
plt.ylabel('Income')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=data_50['Education'], y=data_50['How many books did you read during last 12months?'], palette = 'Set2') #Used ChatGPT to figure out how to change color
plt.title('Books Read by Education Level for 50-Year-Olds')
plt.xlabel('Education Level')
plt.ylabel('Number of Books Read')
plt.xticks(rotation=70)  
plt.show()
```

Next, I'm filtering the dataset to focus on 25-year-old participants and visualizing their reading habits. I created a scatter plot to examine the relationship between the number of books read and income, and a box plot to explore how reading habits vary across different education levels for this age group. I am doing this because I want to explore how reading habits correlate with different life outcomes across different age groups.

```python
data_25 = data[data['Age'] == 25]

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_25['How many books did you read during last 12months?'], y=data_25['Incomes'], color='teal')
plt.title('Books Read vs Income for 25-Year-Olds')
plt.xlabel('Number of Books Read')
plt.ylabel('Income')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=data_25['Education'], y=data_25['How many books did you read during last 12months?'], palette = 'Set2') #Used ChatGPT to figure out how to change color
plt.title('Books Read by Education Level for 25-Year-Olds')
plt.xlabel('Education Level')
plt.ylabel('Number of Books Read')
plt.xticks(rotation=70)  
plt.show()
```

In this analysis, I created two side-by-side count plots to compare the employment status (employed vs. unemployed) of people who read at least five books versus those who did not read any books.

```python
data_read = data[data['How many books did you read during last 12months?'] > 5]
data_no_read = data[data['How many books did you read during last 12months?'] == 0]

# Set up the figure with two subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot for people who read at least one book (Employed vs Unemployed)
sns.countplot(ax=axes[0], x=data_read['Employement'], palette='Set3')
axes[0].set_title('Employed vs Unemployed: People Who Read at Least Five Books')
axes[0].set_xlabel('Employment Status')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=70)

# Plot for people who did not read any books (Employed vs Unemployed)
sns.countplot(ax=axes[1], x=data_no_read['Employement'], palette='Set3')
axes[1].set_title('Employed vs Unemployed: People Who Did Not Read Any Books')
axes[1].set_xlabel('Employment Status')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=70)

# Get the maximum count across both data subsets
max_count = max(data_read['Employement'].value_counts().max(), data_no_read['Employement'].value_counts().max())

# Set the same y-axis limit for both plots for comparison
axes[0].set_ylim(0, max_count)
axes[1].set_ylim(0, max_count)

# Adjust the layout to make sure everything fits
plt.tight_layout()

# Show the plots
plt.show()
```

Lastly, this analysis compares the marital status of people who read at least five books versus those who did not read any books, focusing on the categories 'Married' and 'Divorced'.

```python
# Filter the data for people who read at least one book and people who did not read any books
data_read = data[data['How many books did you read during last 12months?'] > 5]
data_no_read = data[data['How many books did you read during last 12months?'] == 0]

# Filter out other relationship statuses, keeping only 'Married' and 'Divorced'
data_read_filtered = data_read[data_read['Marital status?'].isin(['Married', 'Divorced'])]
data_no_read_filtered = data_no_read[data_no_read['Marital status?'].isin(['Married', 'Divorced'])]

# Set up the figure for side-by-side plots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot for people who read at least one book (Married vs Divorced)
sns.countplot(ax=axes[0], x=data_read_filtered['Marital status?'], palette='Set2')
axes[0].set_title('Married vs Divorced: People Who Read At Least Five Books')
axes[0].set_xlabel('Marital Status')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Plot for people who did not read any books (Married vs Divorced)
sns.countplot(ax=axes[1], x=data_no_read_filtered['Marital status?'], palette='Set3')
axes[1].set_title('Married vs Divorced: People Who Did Not Read Any Books')
axes[1].set_xlabel('Marital Status')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Scale the y-axes of both plots to be the same
y_max = max(axes[0].get_ylim()[1], axes[1].get_ylim()[1])  # Get the maximum y-axis limit
axes[0].set_ylim(0, y_max)  # Set the same y-axis scale for the first plot
axes[1].set_ylim(0, y_max)  # Set the same y-axis scale for the second plot

# Adjust the layout to ensure the plots fit well
plt.tight_layout()

# Show the plots
plt.show()
```

**Results:**

The analysis of the graphs generated showcased several trends regarding reading habits and life outcomes. For both 25-year-olds and 50-year-olds, a weak positive correlation was observed between the number of books read and income, suggesting that individuals who read more may tend to have higher incomes, although the relationship is not strong. In terms of education, the box plot for 25-year-olds indicated that individuals with higher education levels generally read more books, though there was considerable variation within each group. Employment status data showed that people who read at least five books were more likely to be employed compared to those who did not read, with a significantly higher proportion of employed individuals among readers. When examining marital status, both groups of readers and non-readers were predominantly married, although readers had a slightly higher tendency to be married, while non-readers had a somewhat higher proportion of divorced individuals. Lastly, side-by-side comparisons of employment and marital status for readers versus non-readers highlighted that readers were generally employed and married at higher rates, whereas non-readers showed more evenly distributed employment status and a higher proportion of divorced individuals. These findings suggest that reading habits may be loosely linked to certain life outcomes, such as income, employment, and marital status, although the relationships are relatively weak and require further exploration.

**Synthesis and Discussion:**

The analysis shows that reading habits are only weakly connected to various life outcomes, such as income, employment, and marital status. There was a small positive correlation between the number of books read and income, suggesting that people who read more may earn more, but the link is not strong. Education levels were more clearly tied to reading habits, with higher education generally leading to more reading, which matches what previous research has found.

Readers were also more likely to be employed than non-readers, but the difference isn't strong enough to say that reading directly affects employment status. When it comes to marital status, readers were slightly more likely to be married, but the difference in terms of divorce was small. Non-readers had a more balanced employment status but a higher proportion of divorced individuals.

While these findings suggest that reading is linked to positive life outcomes, there are some important limitations. The dataset only provides a snapshot of data at one point in time, so it's hard to determine cause and effect. Also, factors like income, education, and personal circumstances could affect both reading habits and life outcomes, but these were not considered in the analysis.

Future research could focus on long-term studies to see how reading habits impact life outcomes over time. It would also be helpful to look at what types of books people read, as reading fiction might have different effects compared to reading non-fiction. Finally, expanding the dataset to include other factors like lifestyle or personality traits could give a more complete picture of how reading affects different areas of life.
